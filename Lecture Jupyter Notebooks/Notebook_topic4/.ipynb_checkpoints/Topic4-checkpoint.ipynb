<font size = "8"> <center> Frequency Response and Bode Plots

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from ipywidgets import interact, interact_manual, FloatSlider, fixed
import sympy as sp
import control as ctl
from scipy.signal import lti, lsim
```

<font size = "6"> Phase difference betweeen input and output in LTI system for sinusoidal input for transfer function $H(s) = \frac{K}{\tau s+1}$

```python

# Define the system: Transfer function H(s) = K / (tau s + 1)
K = 3;
tau = 0.3;
num = [K]  # Numerator coefficients
den = [tau, 1]  # Denominator coefficients (s + 1)
system = lti(num, den)

# Time vector
t = np.linspace(0, 10, 1000)  # 10 seconds, 1000 points

# Sinusoidal input: u(t) = sin(2t)
omega = 2  # Frequency of sine wave
u = np.sin(omega * t)

# Compute the system response
t_out, y, _ = lsim(system, U=u, T=t)

# Find the first zero-crossing points of input and output
def find_zero_crossing(signal, t):
    """ Find the first zero-crossing (upward) of a signal. """
    for i in range(1, len(signal)):
        if signal[i-1] < 0 and signal[i] >= 0:  # Upward zero-crossing
            return t[i]
    return None  # If no zero-crossing is found

t_input_zero = find_zero_crossing(u, t)
t_output_zero = find_zero_crossing(y, t_out)

# Compute phase delay and convert to degrees
time_delay = t_output_zero - t_input_zero
phase_shift = (time_delay * omega) * (180 / np.pi)  # Convert to degrees

# Plot input and output signals
plt.figure(figsize=(10, 5))
plt.plot(t, u, 'r--', label='Input: sin(2t)', alpha=0.8)
plt.plot(t_out, y, 'b', label='Output response')

# Mark zero-crossings for phase shift visualization
plt.scatter(t_input_zero, 0, color='red', marker='o', label='Input Zero-Crossing')
plt.scatter(t_output_zero, 0, color='blue', marker='o', label='Output Zero-Crossing')

# Draw an arrow showing phase shift
plt.annotate(
    '', xy=(t_output_zero, 0), xytext=(t_input_zero, 0),
    arrowprops=dict(arrowstyle='<->', color='green', lw=2)
)

# Annotate phase shift value
plt.text(
    (t_input_zero + t_output_zero) / 2, 0.1,
    f'Phase Shift: {phase_shift:.2f}°', color='green', fontsize=12, ha='center'
)

# Formatting the plot
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Sinusoidal Response with Correct Phase Difference Visualization')
plt.legend()
plt.grid()
plt.show()
```

<font size="6"> Asymptotic Bode plot of first order system with transfer function ${G(s) = \frac{1}{\tau s + 1}}$ 


```python
# Define system parameters
K = 100;
tau = 0.1  # Time constant
wc = 1 / tau  # Cutoff frequency (rad/s)

# Define frequency range
wout = np.logspace(-2, 3, 1000)  # From 0.001 to 1000 rad/s

# Compute exact magnitude and phase response
mag = K / np.sqrt(1 + (wout / wc) ** 2)
phase = -np.arctan(wout / wc)

# Compute asymptotic magnitude response
asymp_mag = np.piecewise(wout, [wout <= wc, wout > wc], 
                         [20*np.log10(K), lambda w: 20*np.log10(K)-20 * np.log10(w / wc)])

# Compute asymptotic phase response
asymp_phase = np.piecewise(wout, 
                           [wout <= 0.1 * wc, (wout >= 0.1 * wc) & (wout <= 10 * wc), wout > 10 * wc],
                           [0, lambda w: -45 * (np.log10(w / wc) + 1), -90])

# Plot Magnitude Response
plt.figure(figsize=(8, 5))
plt.semilogx(wout, 20 * np.log10(mag), 'b', linewidth=3, label="Actual Magnitude")
plt.semilogx(wout, asymp_mag, '--r', linewidth=3, label="Asymptotic Approximation")
plt.axvline(x=wc, color='k', linestyle='--', linewidth=1.5, label="Cutoff Frequency")
plt.grid(True, which="both", linestyle="--")
plt.xlabel("Frequency (rad/s)")
plt.ylabel("Magnitude (dB)")
plt.title("Asymptotic Bode Magnitude Plot (First-Order)")
plt.legend()
plt.show()

# Plot Phase Response
plt.figure(figsize=(8, 5))
plt.semilogx(wout, np.degrees(phase), 'b', linewidth=3, label="Actual Phase")
plt.semilogx(wout, asymp_phase, '--r', linewidth=3, label="Asymptotic Approximation")
plt.axvline(x=wc, color='k', linestyle='--', linewidth=1.5, label="Cutoff Frequency")
plt.grid(True, which="both", linestyle="--")
plt.xlabel("Frequency (rad/s)")
plt.ylabel("Phase (degrees)")
plt.title("Asymptotic Bode Phase Plot (First-Order)")
plt.legend()
plt.show()
```

<font size="6"> Asymptotic Bode plot for system with one zero ${G(s) = \tau s + 1}$ 


```python

# Define system parameters
tau = 0.1
wz = 1/tau  # Zero frequency (rad/s)

# Define frequency range
wout = np.logspace(-2, 4, 1000)  # From 0.01 to 10000 rad/s

# Compute exact magnitude and phase response
mag = np.sqrt(1 + (wout / wz) ** 2)
phase = np.arctan(wout / wz)

# Compute asymptotic magnitude response
asymp_mag = np.piecewise(wout, 
                         [wout <= wz, wout > wz], 
                         [0, lambda w: 20 * np.log10(w / wz)])

# Compute asymptotic phase response
asymp_phase = np.piecewise(wout, 
                           [wout <= 0.1 * wz, (wout >= 0.1 * wz) & (wout <= 10 * wz), wout > 10 * wz],
                           [0, lambda w: 45 * (np.log10(w / wz) + 1), 90])

# Plot Magnitude Response
plt.figure(figsize=(8, 5))
plt.semilogx(wout, 20 * np.log10(mag), 'b', linewidth=3, label="Actual Magnitude")
plt.semilogx(wout, asymp_mag, '--r', linewidth=3, label="Asymptotic Approximation")
plt.axvline(x=wz, color='k', linestyle='--', linewidth=1.5, label="Zero Frequency")
plt.grid(True, which="both", linestyle="--")
plt.xlabel("Frequency (rad/s)")
plt.ylabel("Magnitude (dB)")
plt.title("Asymptotic Bode Magnitude Plot (One Zero)")
plt.legend()
plt.show()

# Plot Phase Response
plt.figure(figsize=(8, 5))
plt.semilogx(wout, np.degrees(phase), 'b', linewidth=3, label="Actual Phase")
plt.semilogx(wout, asymp_phase, '--r', linewidth=3, label="Asymptotic Approximation")
plt.axvline(x=wz, color='k', linestyle='--', linewidth=1.5, label="Zero Frequency")
plt.grid(True, which="both", linestyle="--")
plt.xlabel("Frequency (rad/s)")
plt.ylabel("Phase (degrees)")
plt.title("Asymptotic Bode Phase Plot (One Zero)")
plt.legend()
plt.show()
```

<font size = "6"> Bode plot for a second order system ${G(s) = \frac{\omega_n^2}{s^2 + 2\zeta\omega_n + \omega_n^2}}$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define system parameters
wn = 10  # Natural frequency (rad/s)
zeta = .5 # Damping ratio

# Define frequency range
wout = np.logspace(-1, 3, 1000)  # From 1 to 100 rad/s

# Compute exact magnitude and phase response
mag = wn**2 / np.sqrt((wn**2 - wout**2) ** 2 + (2 * zeta * wn * wout) ** 2)
phase = -np.arctan2(2 * zeta * wn * wout, wn**2 - wout**2)

# Compute asymptotic magnitude response
asymp_mag = np.piecewise(wout, 
                         [wout <= wn, wout > wn], 
                         [0, lambda w: -40 * np.log10(w / wn)])

# Compute asymptotic phase response
asymp_phase = np.piecewise(wout, 
                           [wout <= 0.1 * wn, (wout >= 0.1 * wn) & (wout <= 10*wn), wout > 10*wn],
                           [0, lambda w: -90 * (np.log10(w / wn) + 1), -180])

# Plot Magnitude Response
plt.figure(figsize=(8, 5))
plt.semilogx(wout, 20 * np.log10(mag), 'b', linewidth=3, label="Actual Magnitude")
plt.semilogx(wout, asymp_mag, '--r', linewidth=3, label="Asymptotic Approximation")
plt.axvline(x=wn, color='k', linestyle='--', linewidth=1.5, label="Natural Frequency")
plt.grid(True, which="both", linestyle="--")
plt.xlabel("Frequency (rad/s)")
plt.ylabel("Magnitude (dB)")
plt.title("Asymptotic Bode Magnitude Plot (Second-Order)")
plt.legend()
plt.show()

# Plot Phase Response
plt.figure(figsize=(8, 5))
plt.semilogx(wout, np.degrees(phase), 'b', linewidth=3, label="Actual Phase")
plt.semilogx(wout, asymp_phase, '--r', linewidth=3, label="Asymptotic Approximation")
plt.axvline(x=wn, color='k', linestyle='--', linewidth=1.5, label="Natural Frequency")
plt.grid(True, which="both", linestyle="--")
plt.xlabel("Frequency (rad/s)")
plt.ylabel("Phase (degrees)")
plt.title("Asymptotic Bode Phase Plot (Second-Order)")
plt.legend()
plt.show()
```

<font size = "6"> BODE PLOTS using the bode_plot command 

     1. Import control as ctl allows use the control package of python. 
     2. Define transfer functions with numerator and denominator polynomials using ctl.tf. 
     3. ctl.bode_plt command is used to plot the bode plots. 



```python

K = 1; tau = 10;
wn = 10;zeta = 0.5;
G1 = ctl.tf([K], [tau, 1])
G2 = ctl.tf([tau, 1],[1])
G3 = ctl.tf([wn^2],[1,2*zeta*wn, wn^2]);
plt.figure(figsize=(10, 7))
ctl.bode_plot(G1, dB=True, omega_limits=[0.001, 1000])      
#plt.title(f"Bode plot of G1")
plt.xlim(0.1, 1000)
ctl.bode_plot(G2, dB=True, omega_limits=[0.001, 1000])      
#plt.title(f"Bode plot of G2")
#plt.xlim(0.001, 1000)
ctl.bode_plot(G3, dB=True, omega_limits=[0.001, 1000])      
plt.suptitle(r"Bode plot of pole, zero and second order system")
plt.xlim(0.001, 1000)
plt.legend([r'$G_1 = \frac{K}{\tau s +1}$',r'$G_2 = \tau s + 1$',r'$G_3 = \frac{\omega_n^2}{s^2+2\zeta \omega_n + \omega_n^2}$'])
plt.show()
```

<font size = "6"> Comparing bode plot for second order system as $\zeta$ varies $ {H(s) =  \frac{101}{s^2+2s+101}}$


<font size = "6"> Poles = $-1 \pm 10i$


```python

H = ctl.tf([101],[1,2,101]);


# Define Transfer Function G(s) = (s + 2) / (s^2 + 0.5s + 1)

sys = H

# Compute poles and zeros
poles = ctl.pole(sys)
zeros = ctl.zero(sys)

# Plot Pole-Zero Diagram
plt.figure(figsize=(6, 6))
plt.axhline(0, color='k', linestyle='--', linewidth=1)  # x-axis
plt.axvline(0, color='k', linestyle='--', linewidth=1)  # y-axis
plt.grid(True, linestyle='--', alpha=0.6)

# Plot poles (X markers, red)
plt.scatter(poles.real, poles.imag, marker='x', color='r', s=100, label="Poles")

# Plot zeros (O markers, blue)
plt.scatter(zeros.real, zeros.imag, marker='o', facecolors='none', edgecolors='b', s=100, label="Zeros")

# Labels and legend
plt.xlabel("Real Axis")
plt.ylabel("Imaginary Axis")
plt.title(r"Pole-Zero Plot of $\frac{101}{s^2+2s+101}$")
plt.legend()
plt.xlim(-1.5,1.5)
plt.show()
plt.figure()
ctl.bode_plot(H, dB=True, omega_limits=[0.5, 100])   
plt.axvline(x=10, color='r', linestyle='--')
plt.suptitle(r"Bode plot of system $\frac{101}{s^2+2s+101}$")
plt.xlim(0.5, 100)
```

<font size = "6"> Different components of $H(s)$

<font size = "6" > $H_1 = \frac{10}{4} \frac{1}{s+1}$ = $\frac{2.5}{s+1}$

<font size = "6" > $H_2 =  \frac{s+5}{5}$

<font size = "6" > $H_3 =  \frac{20}{s+20}$


```python
s = ctl.TransferFunction.s
sys1 = 2.5/(s+1);
sys2 = (s+5)/5;
sys3 = 20/(s+20);

mag1, phase1, omega1 = ctl.bode_plot(sys1, dB=True,omega_limits=[0.01, 1000], Plot=False)
mag2, phase2, omega2 = ctl.bode_plot(sys2, dB=True, omega_limits=[0.01, 1000], Plot=False)
mag3, phase3, omega3 = ctl.bode_plot(sys3, dB=True, omega_limits=[0.01, 1000], Plot=False)
plt.figure()
plt.semilogx(omega1, 20 * np.log10(mag1), 'b')  # Convert to dB scale
plt.semilogx(omega2, 20 * np.log10(mag2), 'g')
plt.semilogx(omega3, 20 * np.log10(mag3), 'r')
plt.xlabel("Frequency (rad/s)")
plt.ylabel("Magnitude (dB)")
plt.title("Bode Magnitude Plot")
plt.legend([r"$H_1 = \frac{2.5}{s+1}$",r"$H_2 = \frac{s+5}{5}$",r"$H_3 = \frac{20}{s+20}$"]);
plt.grid(True)
plt.show()


```
